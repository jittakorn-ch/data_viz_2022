## ตรวจสอบและรวมข้อมูล covid_data

**ตรวจสอบว่าข้อมูลซ้ำกันหรือไม่**

```python
# หาค่า min max ของ column NO.
def min_max(i):
    a = min(i['No.'])
    b = max(i['No.'])
    print(f'min of NO. is {a}')
    print(f'max of NO. is {b}')
```

```python
min_max(covid_data1)    # ดูค่า min max ของตาราง covid_data1
```

```python
min_max(covid_data2)    # ดูค่า min max ของตาราง covid_data2
```

```python
min_max(covid_data3)    # ดูค่า min max ของตาราง covid_data3
```

```python
min_max(covid_data4)    # ดูค่า min max ของตาราง covid_data4
```

- เนื่องจาก ลักษณะของข้อมูลที่ดูจากค่าในคอลัมน์ NO. (รหัสผู้ป่วย) ตาราง covid_data1 และ covid_data2 จะมีค่าซ้ำกัน

(หรัสที่มากที่สุดใน covid_data1 > รหัสที่น้อยที่สุดใน covid_data2)
